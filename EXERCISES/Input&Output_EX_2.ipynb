```python
import pandas as pd
```

```python
data = {
    'name': ['Apple', 'Banana', 'Grapes', 'Orange', 'Mango'],
    'color': ['Red', 'Yellow', 'Green', 'Orange', 'Yellow'],
    'weight': [150, 120, 50, 130, 200]
}
```

```python
# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)
```

```python
# Display the DataFrame
print("Original Data:")
print(df)
```

```python
# Filter the DataFrame
filtered_df = df[df['weight'] > 100]
```

```python
# Display the filtered DataFrame
print("\nFiltered Data (Fruits weighing more than 100 grams):")
print(filtered_df)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(filtered_df['name'], filtered_df['weight'], color=filtered_df['color'])
plt.xlabel('Fruit Name')
plt.ylabel('Weight (grams)')
plt.title('Weights of Fruits Over 100 Grams')
plt.show()
```
